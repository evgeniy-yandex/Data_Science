### 2.2. Подготовим функции для расчета:

**2.2.1. Функция для расчета метрик:**

```python
def calc_metrics(target_valid, predict, proba_one_valid):
    
    metric_accuracy = accuracy_score(target_valid, predict)
    metric_recall = recall_score(target_valid, predict)
    metric_precision = precision_score(target_valid, predict)
    metric_f1 = f1_score(target_valid, predict)
    metric_roc_auc = roc_auc_score(target_valid, proba_one_valid)
    
    print(" Accuracy:", round(metric_accuracy, 3))
    print(" Полнота: ", round(metric_recall, 3))
    print(" Точность:", round(metric_precision, 3))
    print(" f1-score:", round(metric_f1, 3))
    print(' ROC AUC:  %.3f' % (metric_roc_auc))  
    
    return metric_accuracy, metric_recall, metric_precision, metric_f1, metric_roc_auc
    
#    print("Accuracy:", round(accuracy_score(target_valid, predict), 3))
#    print("Полнота: ",  round(recall_score(target_valid, predict), 3))
#    print("Точность:", round(precision_score(target_valid, predict), 3))
#    print("f1-score:", round(f1_score(target_valid, predict), 3))
#    print('ROC AUC:  %.3f' % (roc_auc_score(target_valid, proba_one_valid)))                       
```

**2.2.2. Функция для модели "Дерево решений":**

```python
def create_my_model_tree(X, y, c_w=None):  # c_w == class_weight='balanced'
    best_f1_model_tree = None
    best_f1_result_tree = 0
    best_f1_depth_tree = 1

    for depth in tqdm(range(1, 13)):
        model_tree = DecisionTreeClassifier(random_state=RANDOM_FIX_NUMBER, 
                                            class_weight=c_w, max_depth=depth)  
        model_tree.fit(X, y) 

        f1_result_tree = f1_score(target_valid, 
                                  model_tree.predict(features_valid))
        if f1_result_tree > best_f1_result_tree:
            best_f1_model_tree = model_tree
            best_f1_result_tree = f1_result_tree
            best_f1_depth_tree = depth


#    print(' Расчет по "f1":') 
    print(' Наилучшая модель "дерево решений":', best_f1_model_tree)
    print(' Значение "f1-мера" лучшей модели "дерево решений":', round(best_f1_result_tree, 3))
    print(' Глубина дерева лучшей модели "дерево решений":', best_f1_depth_tree)
    
    return best_f1_depth_tree
```

**2.2.3. Функция для модели "Cлучайный лес":**

```python
def create_my_model_forest(X, y, c_w=None):  # c_w == class_weight='balanced'
    best_f1_model_forest = None
    best_f1_result_forest = 0
    best_f1_est_forest = 0
    best_f1_depth_forest = 0

    for est in tqdm(range(10, 31)):
        for depth in range (1, 31):
            # обучаем модель с заданным количеством деревьев и глубиной дерева
            model_forest = RandomForestClassifier(random_state=RANDOM_FIX_NUMBER, 
                                                  n_estimators=est, max_depth=depth,
                                                  class_weight=c_w)
            model_forest.fit(X, y)

            predictions_valid = model_forest.predict(features_valid) 
            f1_result_forest = f1_score(target_valid, predictions_valid)
            if f1_result_forest > best_f1_result_forest:
                best_f1_model_forest = model_forest
                best_f1_result_forest = f1_result_forest
                best_f1_est_forest = est    
                best_f1_depth_forest = depth

    print(' Наилучшая модель "случайный лес" на валидационной выборке:', best_f1_model_forest)
    print(' Количество деревьев наилучшей модели "случайный лес" на валидационной выборке:', best_f1_est_forest)
    print(' Глубина дерева наилучшей модели "случайный лес" на валидационной выборке:', best_f1_depth_forest)
    print(' Значение "f1-мера" наилучшей модели на валидационной выборке:', round(best_f1_result_forest, 3))   
    
    return best_f1_depth_forest, best_f1_est_forest
```

```python
def plots_matrix_errors(n, text_title):

    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 2))

    fig.suptitle(text_title, y = 1.2, fontsize = 12, )

    cm1 = confusion_matrix(target_valid, global_pred_valid_tree[n],   labels=global_model_tree[n].classes_)
    cm2 = confusion_matrix(target_valid, global_pred_valid_forest[n], labels=global_model_forest[n].classes_)
    cm3 = confusion_matrix(target_valid, global_pred_valid_regr[n],   labels=global_model_regr[n].classes_)

    disp0 = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=global_model_tree[n].classes_)
    disp1 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=global_model_forest[n].classes_)
    disp2 = ConfusionMatrixDisplay(confusion_matrix=cm3, display_labels=global_model_regr[n].classes_)

    disp0.plot(ax=axes[0])
    disp1.plot(ax=axes[1])
    disp2.plot(ax=axes[2])       

    axes[0].set(title='Дерево решений')
    axes[1].set(title='Cлучайный лес')
    axes[2].set(title='Логистическая регрессия')

    plt.show()

    
text_titles = ['Матрицы ошибок для всех трёх моделей при дисбалансе классов',
               'Матрицы ошибок. Метод "Взвешивания"',
               'Матрицы ошибок. Метод "Увеличение выборки"',
               'Матрицы ошибок. Метод "Уменьшения выборки"']

for i in range(4):
    plots_matrix_errors(i, text_titles[i])
```
