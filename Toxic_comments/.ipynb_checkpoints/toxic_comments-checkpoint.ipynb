#### Лемматизация и очистка текста.

```python
# Для определения тега «part-of-speech» (POS) для слов текста
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Объявим лемматизатор.
wnl = WordNetLemmatizer()

# Создадим дополнительные функции.
def lemmitize_text(text):
    '''
    Функция лемматизации текста.
    '''
#    text = [wnl.lemmatize(word) for word in text]
    text = [wnl.lemmatize(word, get_wordnet_pos(word)) for word in text]
    return text

def clear_text(text):
    '''
    Функция для очистки текста.
    '''
    pattern = r'[a-zA-Z]+'
    text = re.findall(pattern, text)
    text = [word.lower() for word in text if word.lower() not in sw]
    text = lemmitize_text(text)
    return " ".join(text)
```

```python
# Создадим дополнительные функцию.
def calc_len(text):
    '''
    Функция подсчёта длина передаваемого текста.
    '''
    text = text.split() # Делим строку на слова
    return len(text)
    
data['text_len'] = data.lemm_text.progress_apply(calc_len)
data.head()
```

```python
# Создадим дополнительные функцию.
def dublicated_words(text):
    '''
    Функция удаления повторяющихся слов в объекте.
    '''
    words = text.split()
    return " ".join(sorted(set(words), key=words.index))

# Удалим повторяющиеся слова в лемматизированном тексте:
data['lemm_text'] = data['lemm_text'].progress_apply(dublicated_words)

# Проверим удаление повторяющихся слов:
data.sort_values(by='text_len', ascending=False)
```

### Дополнительные функции.

```python
df_results = pd.DataFrame(columns=['name_model', 'best_score_', 
                                   'parameters', 'time, c'])

def model_metrics(name_model, best_score, params, time):
    '''
    Функция добавит в результы в df_results: 
    RMSE, время создания и обучения модели
    '''   
    if name_model not in list(df_results['name_model']):  
        df_results.loc[len(df_results.index )] = \
            [name_model, best_score, params, time]
        
    display(df_results)                                   
                                   

def Model_Creation(classificator, parametrs, name):
    '''
    Функция для:
    - создания модели передаваемомго classificator,
    - обучения этой модели по feature_train, target_train,
    - получения F1 метрики best_score_.
    '''
    # Чтобы видеть, что процесс выполняется, поставим tqdm с разовым циклом:
    for _ in tqdm(range(1)): 
        begin = time.time()
        #Применяем GridSearchCV с кросс-валидацией
        grid = GridSearchCV(classificator, parametrs, scoring='f1', 
                            cv=3, return_train_score=True)
        grid.fit(feature_train, target_train)  
        model_metrics(name, grid.best_score_, 
                      grid.best_params_, time.time() - begin)
```

### Классификатор `Логистическая регрессия` с использованием GridSearch.

```python
regression = LogisticRegression(fit_intercept=True, 
                                class_weight='balanced', 
                                random_state=RANDOM_FIX_NUMBER,
                                solver='liblinear'
                                )
regression_parameters = {'C': [1, 10]}   
Model_Creation(regression, regression_parameters, 'LogisticRegression')
```

### Классификатор `Дерева решений` DecisionTreeClassifier с использованием GridSearch.

```python
decisiontree = DecisionTreeClassifier(class_weight='balanced', 
                                      random_state=RANDOM_FIX_NUMBER
                                      )
decisiontree_parameters = {'max_depth': [30]}
    
Model_Creation(decisiontree, decisiontree_parameters, 'DecisionTreeClassifier')
```

### Классификатор `LGBMClassifier`.

```python
lgbm = LGBMClassifier(learning_rate=0.15, 
                      random_state=RANDOM_FIX_NUMBER
                      )
lgbm_parameters = {'max_depth': [30],
                   'n_estimators': [600]
                  }
    
Model_Creation(lgbm, lgbm_parameters, 'LGBMClassifier')
```

### Классификатор `XGBoost` с использованием GridSearch.

```python
xg_cl = xgb.XGBClassifier(random_state=RANDOM_FIX_NUMBER)

xg_parameters = {'max_depth': [30],
                 'n_estimators': [100],
                 'eta': [0.3]
                 }
    
Model_Creation(xg_cl, xg_parameters, 'XGBClassifier')
```
