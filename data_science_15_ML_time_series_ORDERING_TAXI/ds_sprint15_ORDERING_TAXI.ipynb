### Тренды и сезонность.

```python
# Разобъём наши данные на три составляющие: тренд, сезонность и остаток
decomposed = seasonal_decompose(data) 

plt.figure(figsize=(8, 12))

plt.subplot(411)
data.plot(ax=plt.gca())
plt.title("Исходный временной ряд количества заказов такси.", color = 'blue')
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Кол-во заказов.', color = 'blue');

plt.subplot(412)
decomposed.trend.plot(ax=plt.gca())
plt.title("Трендовая составляющая временного ряда.", color = 'blue')
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Значение тренда.', color = 'blue');

plt.subplot(413)
decomposed.seasonal.plot(ax=plt.gca())
plt.title("Сезонная составляющая временного ряда.", color = 'blue')
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Значение периодическое.', color = 'blue');

plt.subplot(414)
decomposed.resid.plot(ax=plt.gca())
plt.title("Остаток декомпозиции временного ряда.", color = 'blue')
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Значение остатка декомпозиции.', color = 'blue');

plt.tight_layout()
```

Отмечаем:
* тренд к увеличению заказ такси по результатам полугода,
* сезонность надо посмотреть более тщательно,
* остатки необъяснимы.

```python
# Изучим сезонную составляющую за неделю
decomposed.seasonal['2018-03-01':'2018-03-07'].plot(figsize=(8, 4));
plt.title("Сезонная составляющая временного ряда за недельный период.", 
          color = 'blue')
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Значение периодическое.', color = 'blue');
```

```python
# Сезонная составляющая за два дня
decomposed.seasonal['2018-03-01':'2018-03-02'].plot(figsize=(8, 4));
plt.title("Сезонная составляющая временного ряда за двухдневный период период.", 
          color = 'blue')
plt.xlabel('Дата и время заказа такси.', color = 'blue')
plt.ylabel('Значение периодическое.', color = 'blue');
```

### Оценка стационарности временного ряда.

`Стационарность` (stationarity) временного ряда означает, что такие компоненты как `тренд` и `сезонность` отсутствуют. Говоря более точно, среднее значение и дисперсия не меняются со смещением во времени.

#### Метод `скользящего среднего`.

```python
# Зададим признак для рассчитываемых тестовых значений скользящего среднего:
data['mean'] = data.rolling(window = 24*7).mean()

plt.figure(figsize = (8,4))
 
# поочередно зададим кривые (перевозки и скользящее среднее) с подписями и цветом
plt.plot(data['num_orders'], label = 'Кол-во заказов такси по часам', color = 'steelblue')
plt.plot(data['mean'], label = 'Скользящее среднее за неделю', color = 'orange')
 
# добавим легенду, ее положение на графике и размер шрифта
plt.legend(title = 'Графики:', loc = 'upper left')
 
# добавим подписи к осям и заголовки
plt.xlabel('Дата заказа такси.', color = 'blue')
plt.ylabel('Кол-во заказов.', color = 'blue')
plt.title('Заказы такси с марта по август 2018.', color = 'blue', fontsize = 12);
 
# выведем обе кривые на одном графике
plt.show()

# Удалим тестовый признак:
data = data.drop(columns=['mean'])
```

`Нестационарность` процесса оценивается визуально. Датасет о заказах такси в аэропрт демонстрирует очевидный тренд - у него растёт среднее значение.

#### Тест `Дики-Фуллера` (Dickey-Fuller test).

Тест заключается в том, что нужно выполнить статистическую проверку следующей гипотезы:
* Нулевая гипотеза предполагает, что процесс нестационарный.
* Альтернативная гипотеза соответственно говорит об обратном. 

Используем пороговое значение, равное 0.05 (5%).

```python
# пеердадим функции 'adfuller' данные о количестве заказо такси и
# сохраним результат в переменной adf_test:
adf_test = adfuller(data['num_orders'])  
    
print('Критерий ADF: %f' % adf_test[0])    
    
# выведем значение p-value:
print('P-значение: %f' % adf_test[1])  

print('Критические значения:')
for key, value in adf_test[4].items():
    print('\t%s: %.3f' % (key, value))

if adf_test[0] < adf_test[4]["5%"]:
    print ("Нулевая гипотеза отвергнута – Временной ряд стационарен")
else:
    print ("Нулевая гипотеза не отвергнута – Временной ряд не стационарен")    
```

```python
# пеердадим функции 'adfuller' данные о количестве заказо такси и
# сохраним результат в переменной adf_test:
adf_test = adfuller(data['num_orders'])  
    
print('Критерий ADF: %f' % adf_test[0])    
    
# выведем значение p-value:
print('P-значение: %f' % adf_test[1])  

print('Критические значения:')
for key, value in adf_test[4].items():
    print('\t%s: %.3f' % (key, value))

if adf_test[0] < adf_test[4]["1%"]:
    print ("Нулевая гипотеза отвергнута – Временной ряд стационарен")
else:
    print ("Нулевая гипотеза не отвергнута – Временной ряд не стационарен")    
```

## Обучение моделей.

### Создадим новые признаки:
* час (так как наблюдаем суточную сезонность)
* лаги(отставания)
* скользящее среднее

```python
def make_features(data, max_lag, rolling_size):
    '''
    Функция для создания следующих дополнительных признаков:
    - час;
    - отстающие значения, максимальный размер которых задается
      параметром max_lag. 
      Такие признаки названы: 'lag_1', 'lag_2' ... до величины max_lag.
    - Скользящее среднее 'rolling_mean', рассчитываемый по аргументу rolling_size
    '''
    data['hour'] = data.index.hour
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_size).mean()
    
make_features(data, 24, 6)

# Проверим наш датасет
data.head()
```

```python
# Удалим строки с пропусками:
data = data.dropna()
# Проверим, что объекты с пропусками удалены
data.head()
```

### Разделим исходные данные на обучающую, валидационную и тестовую выборки.¶

Разбиение на выборки нужно провести без перемешивания данных, сохраняя их хронологический порядок, поэтому воспользуемся функцией `train_test_split` с опцией `shuffle=False`.

В соответствии с заданием, делаем тестовую выборку размером 10% от исходных данных.

```python
#  выделяем тестовую выборку в размере 10% от исходного датасет:
train_valid, test = train_test_split(data, shuffle=False, test_size=0.1)

# БОльшую часть делим на train и valid: 
train, valid = train_test_split(train_valid, shuffle=False, train_size=0.95)

# Проверим получившиеся размеры всех трёх датасетов:
print('Сравним сумму объектов всех трех выборок с размером исходной выборки:', data.shape[0])
print('Сумма размеров обучающей, валидационной и тестовой выборок', 
      train.shape[0] + valid.shape[0] + test.shape[0])
if( data.shape[0] == train.shape[0] + valid.shape[0] + test.shape[0]):
    print('Разделение по выборкам прошло без потерь.')
else:     
    print('Есть потери при разделении на выборки!')      

# Проверим последовательность выборок по времени:
print('\nПоследовательность выборок по времени:')
print(train.index.min(), train.index.max())
print(valid.index.min(), valid.index.max())
print(test.index.min(), test.index.max())
```

Отмечаем правильность проведения разбиения на выборки:
* потерь данных нет,
* данные расположены в хронологическом порядке: валидационные после обучающих, а тестовые после валидационных.

```python
# Поделим все выборки на обучающие и целевые признаки:
features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']

features_valid = valid.drop(['num_orders'], axis=1)
target_valid = valid['num_orders']

features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']

print(features_train.shape, features_train.index.min(), features_train.index.max())
print(features_valid.shape, features_valid.index.min(), features_valid.index.max())
print(features_test.shape, features_test.index.min(), features_test.index.max())
```

Отмечаем правильность проведения деления по признакам:
* потерь данных нет,
* данные расположены в хронологическом порядке: валидационные после обучающих, а тестовые после валидационных.

### Вспомогательные функции.

#### Датафрейм для сохранения результатов работы

```python
df_results = pd.DataFrame(columns=['name_model', 'RMSE_valid', 
                                   'fitting time, с', 'parameters'])

# Создадим список создаваемых моделей:
list_my_models = []

# Настроим, чтобы числа в датасетах отражались с 3 знаками после запятой.
pd.set_option('display.float_format', '{:.3f}'.format) 


def model_metrics(name_model, RMSE, working_time, params):
    '''
    Функция добавит в результы в df_results: 
    RMSE, время создания и обучения модели
    '''   
    if name_model not in list(df_results['name_model']):  
        df_results.loc[len(df_results.index )] = \
            [name_model, RMSE, working_time, params]
        
    display(df_results)
```

#### Функции для создания, обучения модели, расчета метрики RMSE.

Для оценки качества моделей применяем метрику RMSE.

```python
def Creation_model(model, name_model, params):
    '''
    Функция обучает модель и фиксирует время на проведение обучения.
    '''
    # Чтобы видеть, что процесс выполняется, поставим tqdm с разовым циклом:
    for _ in tqdm(range(1)):
        begin = time.time()
        model.fit(features_train, target_train)
        predict_data = model.predict(features_valid)
        RMSE = np.sqrt(mean_squared_error(target_valid, predict_data))      
        model_metrics(name_model, RMSE, time.time() - begin, params) 
        list_my_models.append(model)
    
    return model
```
